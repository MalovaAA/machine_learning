# Преобразование данных

При удалении столбцов не было выявлено сильного улучшения предсказания (можно оставить только G1 и G2 , но в этом мало смысла), поэтому оставим все признаки.

0 <= G < 5 = 1

5 <= G < 10 = 2

10 <= G < 15 = 3

G >= 15 = 4

```python
data_describe = data.describe(include = [object])
data.loc[(data['G3'] >= 0) & (data['G3'] < 5), 'G3'] = 1
data.loc[(data['G3'] >= 5) & (data['G3'] < 10), 'G3'] = 2
data.loc[(data['G3'] <15 ) & (data['G3'] >= 10), 'G3'] = 3
data.loc[data['G3'] >= 15, 'G3'] = 4


data.loc[(data['G2'] >= 0) & (data['G2'] < 5), 'G2'] = 1
data.loc[(data['G2'] >= 5) & (data['G2'] < 10), 'G2'] = 2
data.loc[(data['G2'] <15 ) & (data['G2'] >= 10), 'G2'] = 3
data.loc[data['G2'] >= 15, 'G2'] = 4

data.loc[(data['G1'] >= 0) & (data['G1'] < 5), 'G1'] = 1
data.loc[(data['G1'] >= 5) & (data['G1'] < 10), 'G1'] = 2
data.loc[(data['G1'] < 15 ) & (data['G1'] >= 10), 'G1'] = 3
data.loc[data['G1'] >= 15, 'G1'] = 4


data.head()
```

# Метод К ближайщих соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)

knn.fit(X_train, y_train)

```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

как и ожидалось ошибки достаточно большие, а также - переобучение, так как на тестовой выборке ошибка больше чем на обучающей.

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

подбор парметров дал наилучшее число соседей равное 5 , ошибки найдены ниже.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(X_train, y_train)
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Ошибки удалось исправить, но все еще присутсувует переобучение.

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_predict, y_test))
```

Попробуем подобрать гиперпараметр на одном слое



# Подбор гиперпарамтров классификатора

```python
alpha_arr = np.logspace(-3, 1, 40)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'test')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'train')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

минимальное значение ошибок на обущающей и тестовой выборках

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

получим оптимальное значение параметра регуляризации alpha

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
```

# Классификатор при найденном оптимальном значении.

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

В результате видим что от переобучения избавиться не получилось.

## Метод Random Forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

Построим зависимость числа ошибок на тестовой выборке в зависимости отчисла деревьев

```python
trees_arr = np.arange(50, 500, 10)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    rf = ensemble.RandomForestClassifier(bootstrap=True, n_estimators = trees, oob_score=True)
    rf.fit(X_train, y_train)
    train_err.append(np.mean(y_train != rf.predict(X_train)))
    test_err.append(np.mean(y_test  != rf.predict(X_test)))
    oob_err.append(1 - rf.oob_score_)
    
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. Trees')
plt.xlabel('Trees')
plt.ylabel('Error')
plt.legend()
pass
```

Построим OOB-ошибку в зависимости от числа деревьев.

```python
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlabel('Trees')
plt.ylabel('Error')
plt.title('Error vs. Trees')
plt.legend()
pass
```

Подбор оптимального числа деревьев не дал результата, переобучение все равно присутствует при любом числе деревьев.

Посмотрим на минимальную ошибку:

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Оптимальное число деревьев

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt[0])
```

## Метод Extremely Random Forest

```python
from sklearn import ensemble
ert = ensemble.ExtraTreesClassifier(n_estimators = 100)
ert.fit(X_train, y_train)

err_train = np.mean(y_train != ert.predict(X_train))
err_test  = np.mean(y_test  != ert.predict(X_test))

print(err_train, err_test)
```

Снова видим переобучение.

Построим графики ошибок на тестовой выборке, обучающей выборке и OOB-ошибку в зависимости от числа деревьев.

```python
trees_arr = np.arange(50, 500, 10)
test_err = []
train_err = []
oob_err = []
for trees in trees_arr:
    ert = ensemble.ExtraTreesClassifier(n_estimators = trees, oob_score=True, bootstrap=True)
    ert.fit(X_train, y_train)
    train_err.append(np.mean(y_train != ert.predict(X_train)))
    test_err.append(np.mean(y_test  != ert.predict(X_test)))
    oob_err.append(1 - ert.oob_score_)
plt.plot(trees_arr, train_err, 'b-o', label = 'train')
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(trees_arr), np.min(trees_arr)])
plt.title('Error vs. Trees')
plt.xlabel('Trees')
plt.ylabel('Error')
plt.legend()
pass
```

```python
plt.plot(trees_arr, test_err, 'r-o', label = 'test')
plt.plot(trees_arr, oob_err, 'g-o', label = 'oob')
plt.xlabel('Trees')
plt.ylabel('Error')
plt.title('Error vs. Trees')
plt.legend()
pass
```

Минимальная ошибка:

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

При числе деревьев:

```python
trees_opt = trees_arr[test_err == min_test_err]
print(trees_opt[0])
```

Использование Extremely Random Forest дало результаты хуже чем Random Forest
